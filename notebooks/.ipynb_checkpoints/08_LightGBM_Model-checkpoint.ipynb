# LightGBM Model

```python
#Libraries
import pandas as pd
import pickle
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report
from sklearn.feature_selection import SelectFromModel, RFE
from sklearn import linear_model
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE 
from sklearn.svm import SVC
import lightgbm as ltb
import scikitplot as skplt
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay
```

```python
#Function to evaluate the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
    print('Confusion matrix: \n{}\n'.format(confusion_matrix(ytest, ypred)))
```

```python
#Data
X_train = pd.read_parquet("../data/processed/X_train.parquet")
y_train = pd.read_parquet("../data/processed/y_train.parquet")['isfraud']
X_test = pd.read_parquet("../data/processed/X_test.parquet")
y_test = pd.read_parquet("../data/processed/y_test.parquet")['isfraud']
X_train_scaled = pd.read_parquet("../data/processed/X_train_scaled.parquet")
X_test_scaled = pd.read_parquet("../data/processed/X_test_scaled.parquet")
```

```python
#Classifier
classifier = ltb.LGBMClassifier(n_jobs=2, random_state=0, objective='Binary')
```

```python
#Model
clf_lbt = Pipeline(steps=[
    ('clasificador', classifier)])
```

```python
#Train
clf_lbt.fit(X_train_scaled, y_train)
```

```python
#Predictions
y_pred_lgbm = clf_lbt.predict(X_test_scaled)
ypred_proba_lgbm = clf_lbt.predict_proba(X_test_scaled)
evaluate_model(y_test, y_pred_lgbm, ypred_proba_lgbm)
```

The precision refers to the second column of the confusion matrix and its precision is 1 (very high) for cases 0 and 0.22 cases 1 because there are 244 cases that are 0 and it identifies as 1 and 68 cases that are 1 and it does identify as fraud.

The recall refers to the second row of the confusion matrix. It is 1 for the 0 cases and 0.30 for the 1 cases. The model predicts 160 that should be 1 but predicts them as 0, hits 68. Now it predicts fraud better but misses plus 0.

The f1 is low again, the model is not very good. The accuracy is high but the same thing is still happening as in the base model.

```python
skplt.metrics.plot_cumulative_gain(y_test, ypred_proba_lgbm)
plt.show()
```

The learning of class 1 is worse, it hits very few. However, it is still not a very valid graph for this model due to the imbalance.

```python
skplt.metrics.plot_lift_curve(y_test, ypred_proba_lgbm)
plt.show()
```

The curve goes up and down very sharply, it is not a good model.

```python
skplt.metrics.plot_cumulative_gain(y_test, ypred_proba_lgbm)
plt.show()
```

Similar to the ROC, we see that it is not a good model.

```python
features = [0, 1, (0, 1)]
PartialDependenceDisplay.from_estimator(clf_lbt, X_test_scaled, features)
```

The coefficients shown in the graph are very low, the model is not good.
