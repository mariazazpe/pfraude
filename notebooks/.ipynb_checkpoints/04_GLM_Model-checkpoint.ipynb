# GLM Model

GLM Model are Generalized Linear Models and and mainly includes regression models. In first place, we are going to use 
LinearRegression but we know that it isn't a suitable model for this prediction but we can see what it predicts and later, we use another regression model.

```python
#Libraries
import pandas as pd
import pickle
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report, balanced_accuracy_score
from sklearn.feature_selection import SelectFromModel, RFE
from sklearn import linear_model
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import Lasso, LogisticRegressionCV
from sklearn.linear_model import LinearRegression
#from sklearn.pipeline import Pipeline
import scikitplot as skplt
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay
```

```python
#Function to evaluate the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
    print('Confusion matrix: \n{}\n'.format(confusion_matrix(ytest, ypred)))
```

```python
#we load the data
X_train = pd.read_parquet("../data/processed/X_train.parquet")
y_train = pd.read_parquet("../data/processed/y_train.parquet")['isfraud']
X_test = pd.read_parquet("../data/processed/X_test.parquet")
y_test = pd.read_parquet("../data/processed/y_test.parquet")['isfraud']
X_train_scaled = pd.read_parquet("../data/processed/X_train_scaled.parquet")
X_test_scaled = pd.read_parquet("../data/processed/X_test_scaled.parquet")
```

```python
#We load preprocessor
preprocessor = pickle.load(open('../models/preprocessor.pickle', 'rb'))

#We create smote to reduce the unbalanced data
smt = SMOTE(sampling_strategy=0.5, n_jobs=-1)
```

```python
#Model
clf_lr = Pipeline(steps=[
    #('preprocesador', preprocessor),
    ('clasificador', LinearRegression())
])
```

```python
#We train the model
clf_lr.fit(X_train_scaled, y_train)
```

```python
#Predictions
y_pred_lr = clf_lr.predict(X_test_scaled)
y_pred_lr
```

The prediction is not an array with 0 and 1. This model isn't appropiate for classification problems, for this reason its prediction is useful for us. 

Now, we will use LogisticRegression.

```python
#Model
clf_glm = Pipeline(steps=[
    #('preprocesador', preprocessor),
    ('smote', smt),
    ('clasificador', LogisticRegressionCV(cv=5, n_jobs=2, penalty='l2', random_state=0))])
```

```python
#Train
clf_glm.fit(X_train_scaled, y_train)
```

```python
#Prediction
y_pred_glm = clf_glm.predict(X_test_scaled)
ypred_proba_glm = clf_glm.predict_proba(X_test_scaled)
evaluate_model(y_test, y_pred_glm, ypred_proba_glm)
```

The precision refers to the second column of the confusion matrix and its precision is 1 (very high) for cases 0 and 0.04 for cases 1 because there are 5678 cases that are 0 and identifies as 1 and 207 cases that are 1 and it does identify as fraud.

The recall refers to the second row of the confusion matrix. It is 1 for 0 cases and 0.04 for 1 cases. The model predicts 21 that should be 1 but predicts them as 0, hits 207. It now predicts fraud better but misses plus 0.

The f1 is extremely low. The accuracy is high but the same thing is still happening as in the base model.

```python
# keep probabilities for the positive outcome only
yhat = ypred_proba_glm[:, 1]
# calculate roc curves
fpr, tpr, thresholds = roc_curve(y_test, yhat)
# plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Random Forest')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

The ROC curve is not very suitable to assess this model because of the unbalanced data and it could seem like a good model due to the curve, but it is not.

```python
skplt.metrics.plot_lift_curve(y_test, ypred_proba_glm)
plt.show()
```

In this curve it is better appreciated that the model is not good, the learning of class 1 drops to 0, which indicates that it does not predict very well.

```python
skplt.metrics.plot_cumulative_gain(y_test, ypred_proba_glm)
plt.show()
```

This curve shows that the results of the predictive model are better than random.

```python
features = [0, 1, (0, 1)]
PartialDependenceDisplay.from_estimator(clf_glm, X_test_scaled, features)
```

The third graph indicates the dependence of the variables and as can be seen it is very low, the model does not manage to predict well with these variables.
