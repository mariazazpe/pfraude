# Lasso Model

```python
#Libraries
import pandas as pd
import pickle
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report
from sklearn.feature_selection import SelectFromModel, RFE
from sklearn.linear_model import Lasso, LogisticRegression
from sklearn.pipeline import Pipeline
import scikitplot as skplt
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay
```

```python
#Function to evalute the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
    print('Confusion matrix: \n{}\n'.format(confusion_matrix(ytest, ypred)))
```

```python
#We load the data
X_train = pd.read_parquet("../data/processed/X_train.parquet")
y_train = pd.read_parquet("../data/processed/y_train.parquet")['isfraud']
X_test = pd.read_parquet("../data/processed/X_test.parquet")
y_test = pd.read_parquet("../data/processed/y_test.parquet")['isfraud']
X_train_scaled = pd.read_parquet("../data/processed/X_train_scaled.parquet")
X_test_scaled = pd.read_parquet("../data/processed/X_test_scaled.parquet")
```

```python
#Model
clf_lasso = Pipeline(steps=[
    ('clasificador', LogisticRegression(C=1.5,random_state=0, n_jobs=1, penalty='l1', solver='liblinear', tol= 0.0005))])
```

```python
#Training the model
clf_lasso.fit(X_train_scaled, y_train)
```

```python
#Predictions
y_pred_lasso = clf_lasso.predict(X_test_scaled)
y_pred_proba_lasso = clf_lasso.predict_proba(X_test_scaled)

#Evalution of the model
evaluate_model(y_test, y_pred_lasso, y_pred_proba_lasso)
```

The quality of the model is better than the base model, it identifies fraud but it still fails.

The precision refers to the second column of the confusion matrix and its precision is 1 (very high) for cases 0 and 0.86 for cases 1 because there are 16 cases that are 0 and identifies as 1 and 97 cases that are 1 and it does identify as fraud.

The recall refers to the second row of the confusion matrix. It is 1 for 0 cases and 0.43 for 1 cases. It predicts 131 numbers that should be 1 and predicts as 0, but hits 97. The prediction is better.

The f1 has also improved. The accuracy is high but the same thing is still happening as in the base model.

```python
# keep probabilities for the positive outcome only
yhat = y_pred_proba_lasso[:, 1]
# calculate roc curves
fpr, tpr, thresholds = roc_curve(y_test, yhat)
# plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Random Forest')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

The ROC curve is not very suitable to assess this model because of the unbalanced data and it could seem like a good model due to the curve, but it is not.

```python
skplt.metrics.plot_lift_curve(y_test, y_pred_proba_lasso)
plt.show()
```

In this curve it is better appreciated that the model is not good, the learning of class 1 drops to 0, which indicates that it does not predict very well.

```python
skplt.metrics.plot_cumulative_gain(y_test, y_pred_proba_lasso)
plt.show()
```

This curve shows that the results of the predictive model are better than random.

```python
features = [0, 1, (0, 1)]
PartialDependenceDisplay.from_estimator(clf_lasso, X_test_scaled, features)
```

The third graph indicates the dependence of the variables and as can be seen it is very low, the model does not manage to predict well with these variables.
